## Начинаем работу с колонками

```python
df = pd.read_csv('/home/cdxxi/Загрузки/_data.csv')
```

```python
#Выкидываем ненужные колонки
df = df.drop(columns=['Серия дома', 'Ссылка на объявление', 'Телефоны', 'Описание', 'Unnamed: 0', 'Название ЖК', 'Можно с детьми/животными', 'Тип'])
```

```python
#Исправили цену (теперь только цена/мес)
def price(s):
    return int(s.split('.')[0])

df['Цена'] = df['Цена'].apply(price)
```

```python
#Удалим дубликаты по ID
df.drop_duplicates(subset=['ID  объявления'], keep='first', inplace=True)
```

```python
#Удаляем все, кроме Москвы
df = df[df['Адрес'].str.contains('Москва')].reset_index()
```

## Работаем с пропусками

```python
#Сколько NaN и где? (проверяю по мере заполнения)
df.isna().sum()
```

```python
#Помучаемся с адресами... Убираем шоссе оттуда, где оно есть
df['Адрес'] = [x.split('ш. ')[0] for x in df['Адрес']]
```

```python
#Создаем список всех возможных улиц для редактирования
lst = df['Адрес'].tolist()
```

```python
#Манипуляции с адресами..
lst1 = [str(index) + ', ' + features for index, features in enumerate(lst)]
```

```python
lst2 = [x.split(',') for x in lst1]
lst3 = [' '.join(' '.join(x).strip().split()[2:-1]) for x in lst2]

#Обновляем колонку в df
df['Адрес'] = lst3
```

```python
#Парковка (заменяем NaN)
df['Парковка'] = df['Парковка'].fillna('бесплатная во дворе')
```

```python
#Количество комнат (если длина строки больше 1 - значит там не только значение, вытаскиваем его по индексу)
def room_count(s):
    if len(s) > 1:
        return int(s.split(',')[0])
    else:
        return int(s)
df['Количество комнат'] = df['Количество комнат'].fillna('0')
df['Количество комнат'] = df['Количество комнат'].apply(room_count)
```

```python
#Метро (заполняем NaN)
df['Метро'] = df['Метро'].fillna('м. Отсутствует рядом (0 мин пешком)')
df['Метро'] = [x.split(' (')[0] for x in df['Метро']]
df['Метро'] = [x.split('. ')[1] for x in df['Метро']]
```

```python
df['Ремонт'] = df['Ремонт'].fillna('Без ремонта')
```

```python
#Площадь - заполним пробелы
df['Площадь общая, м2'] = [float(x.split('/')[0]) for x in df['Площадь, м2']]
```

```python
#Выкинем теперь площадь комнат
df = df.drop(columns=['Площадь комнат, м2', 'Площадь, м2'])
```

```python
#Балконы - заполним NaN, разделим по двум столбцам
df['Балкон'] = df['Балкон'].fillna('Балкон (0)')
df['Лоджия'] = [x.split(' ')[-1] if x.split(' ')[-2] == 'Лоджия' else 0 for x in df['Балкон']]
df['Балкон'] = [x.split(' ')[1] if x.split(' ')[0] == 'Балкон' else 0 for x in df['Балкон']]
```

```python
#Убрать скобки у балконов
df['Балкон'] = [int(x[1]) if type(x) == str else 0 for x in df['Балкон']]
df['Лоджия'] = [int(x[1]) if type(x) == str else 0 for x in df['Лоджия']]
```

```python
#Окна - заполним NaN 
df['Окна'] = df['Окна'].fillna('На улицу или двор')
```

```python
#Санузлы - делаем как с балконами
df['Санузел'] = df['Санузел'].fillna('Совмещенный (1)')
df['Санузел разд.'] = [x.split(' ')[-1] if x.split(' ')[-2] == 'Раздельный' else 0 for x in df['Санузел']]
df['Санузел совм.'] = [x.split(' ')[1] if x.split(' ')[0] == 'Совмещенный' else 0 for x in df['Санузел']]
```

```python
#Вторая часть про санузлы - вытащили значения, сделали из них int принудительно :)
df['Санузел разд.'] = [int(x[1]) if type(x) == str else 0 for x in df['Санузел разд.']]
df['Санузел совм.'] = [int(x[1]) if type(x) == str else 0 for x in df['Санузел совм.']]
df = df.drop(columns=['Санузел'])
```

```python
#Разберемся с колонкой "Дополнительно". Логика такая -> не прописано - значит квартира пустая
df['Дополнительно'] = df['Дополнительно'].fillna('Отсутствует')
```

```python
#Сделаем в колонке два вывода - "Присутствуют удобства" если там все, кроме "Отсутствует", и "Отсутствуют удобства" в ином случае
df['Дополнительно'] = ['Отсутствуют удобства' if x == 'Отсутствует' else 'Присутствуют удобства' for x in df['Дополнительно']]
```

```python
#Потолки... настало ваше время... Пропуски заполним средним значением, округлим после точки до двух символов
#Предварительно переведем см в м (надеемся, что трехзначные числа - это см)
df['Высота потолков, м'] = [x / 100 if x > 30 else x for x in df['Высота потолков, м']]
df['Высота потолков, м'] = [x / 10 if x > 20 else x for x in df['Высота потолков, м']]
df['Высота потолков, м'] = df['Высота потолков, м'].fillna(round(df['Высота потолков, м'].mean(), 2))
```

```python
#Из колонки Дом получить колонку Дом и Этаж
df['Тип дома'] = [x.split(', ')[-1] if len(x.split(', ')[-1]) > 6 else 'Неизвестно' for x in df['Дом']]
df['Количество этажей'] = [int(x.split('/')[1].split(',')[0]) for x in df['Дом']]
df['Этаж квартиры'] = [int(x.split('/')[0]) for x in df['Дом']]
df = df.drop(['Дом'], axis=1)
```

```python
#Лифты. Заполним NaN - пусть будет Пасс (0), Груз (0), если меньше 5 этажей
df['Лифт'] = [x if x == x else ['Пасс (1), Груз (1)' if df['Количество этажей'].iloc[i]>5 else 'Пасс (0), Груз (0)'][0]for i, x in enumerate(df['Лифт'])]
```

```python
#Разделим лифты на две колонки - груз и пасс, удалим старую колонку
df['Лифт груз.'] = [x.split(' ')[-1] if x.split(' ')[-2] == 'Груз' else 0 for x in df['Лифт']]
df['Лифт пасс.'] = [x.split(' ')[1] if x.split(' ')[0] == 'Пасс' else 0 for x in df['Лифт']]
```

```python
#Заменим у лифтов значения на type=int + выкинем ненужную колонку
df['Лифт груз.'] = [int(x[1]) if type(x) == str else 0 for x in df['Лифт груз.']]
df['Лифт пасс.'] = [int(x[1]) if type(x) == str else 0 for x in df['Лифт пасс.']]
df = df.drop(columns=['Лифт'])
```

```python
#Обработаем мусоропровод от NaN
df['Мусоропровод'] = df['Мусоропровод'].fillna('Нет')
```

```python
#Поработаем с выбросами
df_v2 = df[df['total_area'] < 600].reset_index(drop = True)
df_v2 = df[df['price'] < 1500000].reset_index(drop = True)
df_v2 = df[df['price'] > 20000].reset_index(drop = True)
```

```python
#Сделаем словари для замены категориальных признаков на числовые (метро и адрес рассмотрим отдельно)
parking = {x: i for i, x in enumerate(sorted(list(set(df['parking']))))}
home_ren = {x: i for i, x in enumerate(sorted(list(set(df['home_renovation']))))}
window = {x: i for i, x in enumerate(sorted(list(set(df['windows']))))}
add_am = {x: i for i, x in enumerate(sorted(list(set(df['additional_amenities']))))}
chute = {x: i for i, x in enumerate(sorted(list(set(df['garbage_chute']))))}
building = {x: i for i, x in enumerate(sorted(list(set(df['building_type']))))}
```

```python
#Заменим данные в таблице
df_v2['parking'] = [parking[x] for x in df_v2['parking']]
df_v2['home_renovation'] = [home_ren[x] for x in df_v2['home_renovation']]
df_v2['windows'] = [window[x] for x in df_v2['windows']]
df_v2['additional_amenities'] = [add_am[x] for x in df_v2['additional_amenities']]
df_v2['garbage_chute'] = [chute[x] for x in df_v2['garbage_chute']]
df_v2['building_type'] = [building[x] for x in df_v2['building_type']]
```

```python
#Функция для кодировки адреса и метро
def cat_code(df, cat, real):
    return df[cat].map(df.groupby(cat)[real].mean())
```

```python
#Кодируем метро
df_v3 = df_v2.copy()
df_v3['subway'] = cat_code(df_v3, 'subway', 'price')
```

```python
#Кодируем адреса
df_v3['address'] = cat_code(df_v3, 'address', 'price')
```

```python
#Используем dummy для лучшего обучения
cat_features = ['parking','home_renovation', 'windows', 'additional_amenities', 'garbage_chute', 'building_type']
features = ['room_number', 'balcony', 'ceiling_height','total_area','living_area','kitchen_area','loggia','separate_bathroom','shared_bathroom','floors_number','apartment_floor','freight_elevator','passenger_elevator']

for f in cat_features:
    df_v3 = pd.concat([df_v3, pd.get_dummies(df_v3[f], prefix = f)], axis = 1)
    features.extend([col for col in df_v3 if col.startswith(f + '_')])
features
```

```python
#Используем One-Hot 
for col in cat_features:
    if df_v3[col].dtype == 'object':
        df_v3[col] = pd.factorize(df_v3[col])[0]
```

```python
#Вместо True False сделаем 1 и 0
def true_or_false(x):
    if x == True:
        return 1
    elif x == False:
        return 0
    else:
        return x

need_to_change = ['parking_0', 'parking_1', 'parking_2', 'parking_3', 'parking_4', 'parking_5',
       'home_renovation_0', 'home_renovation_1', 'home_renovation_2', 'home_renovation_3', 
                  'windows_0', 'windows_1', 'windows_2', 'windows_3',
       'additional_amenities_0', 'additional_amenities_1', 'garbage_chute_0',
       'garbage_chute_1', 'building_type_0', 'building_type_1',
       'building_type_2', 'building_type_3', 'building_type_4',
       'building_type_5', 'building_type_6', 'building_type_7',
       'building_type_8', 'building_type_9']
for i in need_to_change:
    df_v3[i] = df_v3[i].apply(true_or_false)
```
